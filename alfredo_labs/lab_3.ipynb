El error de regimen permanente se define por
$$ε_{ss} = \lim_{n \to \infty}\ ε[n] $$
Utilizando el teorema del valor final (i.e., $x[∞] = \lim_{z→1}(z − 1)X(z)$), encuentre una expresion de $ε_{ss}$ en funcion de $L(z)$ y $Θ(z)$ (suponer $ε[n]=0$, $n<0$).

# Desarrollo

Definimos la secuencia $$\epsilon[n] = \theta[n]-\phi[n]$$
$$ E(z) = \Theta(z)-\Phi(z)$$
Sabemos que la funcion de transferencia del PLL es 
$$H(z)=\frac{G(z)}{1+G(z)},G(z) = L(z)\frac{z^{-1}}{1-z^{-1}}$$ y $$\Phi(z)=E(z)L(z)\frac{z^{-1}}{1-z^{-1}}$$
$$ E(z) = \Theta(z)-E(z)L(z)\frac{z^{-1}}{1-z^{-1}}$$
$$ E(z)\big(1+L(z)\frac{z^{-1}}{1-z^{-1}}\big) = \Theta(z)$$
$$ E(z) = \Theta(z)\big(\frac{1-z^{-1}}{1-z^{-1}+L(z)z^{-1}}\big)$$
Aplicamos el *Teorema del valor final* y obtenemos una expresion del regimen de error permanente
$$ ε_{ss} = \lim_{z→1}(z-1)\Theta(z)\big(\frac{1-z^{-1}}{1-z^{-1}+L(z)z^{-1}}\big)$$







## Demostracion de regimen de error permanente
> Para $θ[n] = Ω_0nu[n]$, demostrar que $ε_{ss}$ es cero si L(z) tiene un polo en z = 1.


$$\Theta(z)=\sum_{n=-\infty}^{\infty}Ω_0nu[n]z^{-n}$$
$$\Theta(z)=\sum_{n=0}^{\infty}Ω_0nz^{-n}=Ω_0\frac{z}{(z-1)^2},|z|>1$$
$$ ε_{ss} = \lim_{z→1}(z-1)Ω_0\frac{z}{(z-1)^2}\big(\frac{1-z^{-1}}{1-z^{-1}+L(z)z^{-1}}\big)$$
$$ ε_{ss} = \lim_{z→1}Ω_0\frac{(z-1)}{(z-1)}\big(\frac{1}{1-z^{-1}+L(z)z^{-1}}\big)$$
$$ ε_{ss} = \lim_{z→1}Ω_0\big(\frac{1}{1-z^{-1}+L(z)z^{-1}}\big)$$

Reemplazo $L(z)=L'(z)/(z-1)$ (funcion de transferencia con un polo en $z=1$)

$$ ε_{ss} = \lim_{z→1}Ω_0\big(\frac{z-1}{L'(z)z^{-1}}\big)$$
$$ ε_{ss} = \lim_{z→1}Ω_0\big(\frac{z^2-z}{L'(z)}\big)=0$$






## Verificacion de funcion de lazo cerrado 
> Sea $L(z) = K_p$ (filtro proporcional). Verifique que la funcion de lazo cerrado es $$H(e^{jΩ}) = \frac{K_p }{e^{jΩ} +K_p −1}$$ con frecuencia de corte (-3dB) dada por $Ω_c ≈ K_p$. 



$$H(z)=\frac{G(z)}{1+G(z)},G(z) = L(z)\frac{z^{-1}}{1-z^{-1}}$$  

$$H(z)=\frac{L(z)\frac{z^{-1}}{1-z^{-1}}}{1+L(z)\frac{z^{-1}}{1-z^{-1}}}$$  

$$H(z)=\frac{K_p}{{z-1}+K_p}$$
Reemplazando $z=e^{j\Omega}$ para obtener la Transformada de Fourier
$$H(e^{j\Omega})=\frac{K_p}{{e^{j\Omega}-1}+K_p}$$



Calculo $||H(e^{j\Omega_c})||^2=0.5$
$$||H(e^{j\Omega_c})||^2=\big\|\frac{K_p}{{\cos(\Omega_c)+j\sin(\Omega_c)-1}+K_p}\big\|^2$$

Para valores de $\Omega_c<<1$ se puede aproximar $\cos(\Omega_c)=1$ y $\sin(\Omega_c)=\Omega_c$. Entonces,

$$||H(e^{j\Omega_c})||^2=\Big\|\frac{K_p}{j\Omega_c+K_p}\Big\|^2$$
$$||H(e^{j\Omega_c})||^2=\frac{||K_p||^2}{||j\Omega_c+K_p||^2}=0.5$$

Claramente observamos que $${2||K_p||^2}={||j\Omega_c+K_p||^2}$$
$${K_p}={\Omega_c}$$






## Analisis de respuesta del lazo de filtros L1 y L2 
> Analice la respuesta del lazo para los siguientes filtros: $$L_1(z) =z^{−D}{K_p}$$ $$L_2(z)=(z^{−D})( K_p +K_i\frac{z^{-1}}{1−z^{−1}})$$ donde D es la latencia del lazo.

$$H(z)=\frac{z^{−D}K_p}{{z-1}+z^{−D}K_p}$$
$$H(z)=\frac{z^{−(D+1)}K_p}{{1-z^{-1}}+z^{−(D+1)}K_p}$$


$$H(e^{j\Omega})=\frac{e^{-j\Omega D}K_p}{{e^{j\Omega}-1}+e^{-j\Omega D}K_p}$$

Ahora con $L_2(z)$
$$H(z)=\frac{(z^{−D})( K_p +K_i\frac{z^{-1}}{1−z^{−1}})\frac{z^{-1}}{1-z^{-1}}}{1+(z^{−D})( K_p +K_i\frac{z^{-1}}{1−z^{−1}})\frac{z^{-1}}{1-z^{-1}}}$$  


$$H(z)=\frac{K_p z^{−(D+1)} +K_i\frac{z^{-(D+2)}}{(1−z^{−1})}}{(1-z^{-1})(1+(z^{−D})( K_p +K_i\frac{z^{-1}}{1−z^{−1}})\frac{z^{-1}}{1-z^{-1}})}$$  


$$H(z)=\frac{K_p z^{−(D+1)} +K_i\frac{z^{-(D+2)}}{(1−z^{−1})}}
{1-z^{-1}+(z^{−D})( K_p +K_i\frac{z^{-1}}{1−z^{−1}})z^{-1}}$$  


$$H(z)=\frac{K_p z^{−(D+1)} +K_i\frac{z^{-(D+2)}}{(1−z^{−1})}}
{1-z^{-1}+z^{−(D+1)}K_p +z^{−(D+1)}K_i\frac{z^{-1}}{1−z^{−1}}}$$  


$$H(z)=\frac{K_p z^{−(D+1)}−K_p z^{−(D+2)} +K_i{z^{-(D+2)}}{}}
{1−z^{−1}-(1−z^{−1})z^{-1}+(1−z^{−1})z^{−(D+1)}K_p +z^{−(D+1)}K_i{z^{-1}}}$$  

$$H(z)=\frac{K_p z^{−(D+1)}−K_p z^{−(D+2)} +K_i{z^{-(D+2)}}{}}
{1−2z^{-1}−z^{−2}+z^{−(D+1)}K_p−z^{−(D+2)}(K_p + K_i)}$$  

A continuacion se plotean las respuestas en frecuencia de $L_1$ y $L_2$, se diseña con ganancias $K_p$ y $K_i$ sean estables para 400KHz.

```python
import zplane
import numpy as np
import numpy as np
import matplotlib.pyplot as pl
from pylab import rcParams
rcParams['figure.figsize'] =  12,8
import matplotlib
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['text.latex.unicode'] = True
matplotlib.rcParams['font.size']=15
import matplotlib.pyplot as pl
```

```python
Bl=.40e6;              # Ancho de banda, en Hz
fs=400e6;              #% Frecuencia de muestreo, in Hz
D=0;                   #% Latencia en ciclos de reloj de muestreo
Kp=Bl/fs*2*np.pi;    #% Ganancia proporcional
Ki=Kp*1e-3*1.0;   #% Ganancia integral
```

```python
omega=np.arange(0.000001*np.pi,np.pi,0.0001*np.pi);
z=[np.exp(np.complex(0,i)) for i in omega];


#% Respuestas de lazo abierto y cerrado
L1z=np.power(z,-D)*Kp;
L2z=np.power(z,-D)*(Kp+(Ki*np.power(z,-1)/(1-np.power(z,-1))));
G1z=L1z*np.power(z,-1)/(1-np.power(z,-1));
G2z=L2z*np.power(z,-1)/(1-np.power(z,-1));
H1z=G1z/(1+G1z);
H2z=G2z/(1+G2z);

#% Funcion de transferencia (en funcion de "f" de tiempo continuo)

pl.semilogx(omega*fs/2/np.pi,20*np.log10(abs(H1z)),'b',label='L1')
pl.semilogx(omega*fs/2/np.pi,20*np.log10(abs(H2z)),'r',label='L2');
pl.xlabel('$f [Hz]$');
pl.ylabel('$|H(j\omega)|$');
pl.legend()
pl.grid(True)

pl.show()
```

## Evalue la establidad del lazo en funcion de la latencia D utilizando el diagrama de polos y ceros 

$$H(z)=\frac{K_p}{{z-1}+K_p}$$
$$H(z)=\frac{z^{−D}K_p}{{z-1}+z^{−D}K_p}$$
$$H(z)=\frac{z^{−(D+1)}K_p}{{1-z^{-1}}+z^{−(D+1)}K_p}$$



```python
def create_pll_L1(D=2,Kp=Kp):
    b=np.zeros(D+1)
    b[D]=Kp
    a=np.zeros(D+2)
    #a[0]=1
    a[0]=1
    if(D==0):
        a[D+1]=-1+Kp
    else:
        a[1]=-1
        a[D+1]=Kp
                
    return b,a


b,a=create_pll_L1(0)

```

```python
b,a=create_pll_L1(0)



r,p,k=zplane.zplane(b,a)
print "Lista de polos fuera del circulo unitario en L1: "+str([abs(i) for i in p if abs(i)>1])
```

```python
b,a=create_pll_L1(50)
r,p,k=zplane.zplane(b,a)
print "D="+str(50)+" Lista de polos fuera del circulo unitario: "
print str([abs(i) for i in p if abs(i)>1])
b,a=create_pll_L1(500)
r,p,k=zplane.zplane(b,a)
print "D= "+str(500)+" Lista de polos fuera del circulo unitario: "
print str([abs(i) for i in p if abs(i)>1])

```

## Estabilidad del lazo

> Seleccione valores de ganancias apropiados (e.g., asegurar estabilidad del lazo) y verifique la magnitud de la respuesta en frecuencia por medio de simulaciones (e.g., usar señales de entrada $θ[n] = A cos (Ωn)$).

```python
f0=.10e6              #Frecuencia de Tono de Jitter
A=.0001  # Amplitud de Tono de Jitter
SNR_TRdB=60.0
SNR_TR=10**(SNR_TRdB/10.0);
Omega_0=2*np.pi*.50; #Offset de frecuencia
TEST_JIT_NOISE=1; # 1: test de jitter senoidal+ruido / 0: test de offset de frecuencia
```

```python
D=5
```

```python
res_jitter=None
def run_tone(d,a,f):
    N=int(1e6)
    pedel=np.zeros(d);
    pe_v=np.zeros(N);
    sum_pe_v=np.zeros(N);
    tita_n_v=np.zeros(N);
    phi_n_v=np.zeros(N);
    noise=np.sqrt(.5)*np.random.randn(N)
    sum_pesd=0;
    phi_n=0;
    for n in range(N):
        tita_n=2*np.pi*a*np.cos(2*np.pi*f*n/fs)
        tita_n=tita_n+noise[n]/np.sqrt(SNR_TR);

        pe=tita_n-phi_n;
        if(D>0):
            pesd=pedel[-1];
            pedel=np.append([pe],pedel[:-2])
        else:
            pesd=pe

        phi_n_v[n]=phi_n;
        c_n=Kp*pesd+Ki*sum_pesd*1.0;
        phi_n=phi_n+c_n;
        sum_pesd=sum_pesd+pesd;
        pe_v[n]=pe;
        sum_pe_v[n]=sum_pesd;
        tita_n_v[n]=tita_n;
    res_jitter=phi_n_v[-50000:]-np.mean(phi_n_v[-50000:]);
    pl.plot([i/fs for i in range(len(phi_n_v))][-7000:],
            phi_n_v[-7000:],label=str(f0)+"Hz")
    return res_jitter
```

Aplico varias sinusoidales sin ruido y con igual amplitud y observo la respuesta en frecuencia por simulación

```python
D=0
f0=1e6              #Frecuencia de Tono de Jitter
A=.001  # Amplitud de Tono de Jitter
run_tone(D,A,f0)

f0=0.5e6              #Frecuencia de Tono de Jitter
run_tone(D,A,f0)


f0=0.1e6              #Frecuencia de Tono de Jitter
run_tone(D,A,f0)


f0=0.050e6              #Frecuencia de Tono de Jitter
run_tone(D,A,f0)

pl.title('Salidas de PLL')
pl.legend()
pl.xlabel('$t$')
pl.ylabel('$\phi_n$')
pl.grid(True)
pl.legend()
pl.show()

```
